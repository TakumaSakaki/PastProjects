```python
import pandas as pd
from filer3 import Filer

data = Filer.read_tsv('train_kwh.tsv')
df = pd.DataFrame(data[1:], columns = data[0])
```

```python
df.head()
```

```python
df['SOLA01'] = df['SOLA01'].map(lambda x: int(x) if x != "" else None)
df['SOLA02'] = df['SOLA02'].map(lambda x: int(x) if x != "" else None)
df['SOLA03'] = df['SOLA03'].map(lambda x: int(x) if x != "" else None)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Year'] = df['datetime'].map(lambda x : x[0:4])
df['Month'] = df['datetime'].map(lambda x : x[4:6])
df['Date'] = df['datetime'].map(lambda x : x[6:8])
df['Time'] = df['datetime'].map(lambda x : x[8:12])
```

```python
#30分刻みへ
df['SOLA01_30'] = df['SOLA01'] + df['SOLA01'].shift(1) + df['SOLA01'].shift(2)
df['SOLA02_30'] = df['SOLA01'] + df['SOLA02'].shift(1) + df['SOLA02'].shift(2)
df['SOLA03_30'] = df['SOLA03'] + df['SOLA03'].shift(1) + df['SOLA03'].shift(2)
```

```python
df[40:60]
```

```python
df = df[df.index % 3 == 2]
```

```python
df['Solar1'] = df['SOLA01_30']
df['Solar2'] = df['SOLA02_30']
df['Solar3'] = df['SOLA03_30']
df = df[['Year', 'Month', 'Date', 'Time', 'Solar1', 'Solar2', 'Solar3']]
df = df.reset_index(drop=True)
```

```python
df['Time'] = df['Time'].map(lambda x : int(x[0:2]) + 0.5 if x[2:4] == '30' else int(x[0:2]))
```

```python
df.to_csv("train.csv", index = False,  header = True)
```
